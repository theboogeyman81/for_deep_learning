Forward_Pass_of_a_Single_Neuron


```python
import numpy as np
```

```python
X = np.array([[0.5],
             [0.1],
             [0.3]])

y = np.array([[1]])
```

```python
weights = np.random.rand(X.shape[0] , 1)
bias = np.random.rand(1)
```

```python
def step_function(z):
  if z >= 0:
    return 1
  else:
    return 0
```

```python
def simple_neuron(X, weights , bias):
  z = np.dot(X.T , weights) + bias

  output = step_function(z)

  if output == 1:
    print('neuron is fired')
  else:
    print('neuron is not fired')
```

```python
simple_neuron(X, weights, bias)
```
