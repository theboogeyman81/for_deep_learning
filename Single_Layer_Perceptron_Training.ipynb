```python
import numpy as np

X = np.array([[0,0] , [0,1] , [1,0] , [1,1]])
y = np.array([[0] , [1] , [1] , [0]])


weights = np.random.rand(X.shape[1] ,1)
bias = np.random.rand(1)


def forward_propagation(X , weights , bias):
  z = np.dot(X , weights) + bias
  output = 1/(1+np.exp(-z))
  return output
```

```python
def loss(y_true , y_pred):
  return np.mean((y_true - y_pred) ** 2)
```

```python
def backward_propagation(X , y , y_pred , weights , learning_rate):
  m = X.shape[0]
  dz = y_pred - y
  dw = np.dot(X.T , dz) / m
  db = np.sum(dz) / m


  weights -= learning_rate * dw
  bias = bias - learning_rate * db
  return weights , bias



def train(X, y , weights , bias , learning_rate=0.1 , epochs=10000):
  for epoch in range(epochs):
    y_pred = forward_propagation(X , weights , bias)
    weights , bias = backward_propagation(X , y , y_pred , weights , learning_rate)

    if epoch % 100 == 0:
      current_loss = loss(y , y_pred)
      print(f"Epoch {epoch} , loss:{current_loss}")
  return weights , bias



```

```python
def predict(X , weights , bias):
  output = forward_propagation(X , weights , bias)
  predictions = (output > 0.5).astype(int)
  return predictions


def accuracy(y_true , y_pred):
  correct_predictions = np.sum(y_true == y_pred)
  total_predictions = len(y_true)
  accuracy = correct_predictions / total_predictions
  return accuracy
```

```python
trained_weights , trained_bias = train(X , y , weights , bias)
predictions = predict(X , trained_weights , trained_bias)
model_accuracy = accuracy(y , predictions)
print(f"Model Accuracy: {model_accuracy * 100}%")
```
